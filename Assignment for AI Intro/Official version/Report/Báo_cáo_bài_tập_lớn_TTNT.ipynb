# **2. Giới thiệu và mô tả bài toán**
# 2.1. Giới thiệu
* Bài toán được giới thiệu là vấn đề về tìm đường trong mê cung sử dụng các thuật toán tìm kiếm đồ thị. Mê cung được mô phỏng thông qua một lưới ô vuông, trong đó có các ô bắt đầu, ô kết thúc, và các ô cản trở. Mục tiêu là tìm đường từ ô bắt đầu đến ô kết thúc, tránh qua các ô cản trở.

# 2.2. Mô tả bài toán
Bài toán tìm đường trong mê cung là một vấn đề quen thuộc trong lĩnh vực trí tuệ nhân tạo và robotica. Trong môi trường mê cung, có một số điểm quan trọng:

* Ô bắt đầu (s): Điểm xuất phát của hành trình.

* Ô kết thúc (e): Điểm mục tiêu, nơi cần đến.

* Ô cản trở (Walls): Các ô không thể đi qua.

* Đường đi (Path): Các ô có thể đi qua để đạt được mục tiêu.

Mô hình mê cung được biểu diễn dưới dạng một lưới ô vuông, và mục tiêu là tìm một đường đi hợp lý từ ô bắt đầu đến ô kết thúc.

# 2.3. Ứng dụng thực tế
Bài toán tìm đường trong mê cung có nhiều ứng dụng trong thực tế:

* Robot di chuyển: Trong robotica, robot cần có khả năng tự định hình đường đi trong môi trường không gian hạn chế.

* Định tuyến mạng lưới: Trong mạng lưới viễn thông, việc định tuyến đường đi hiệu quả giữa các điểm truy cập là quan trọng.

* An toàn và khẩn cấp: Trong ngữ cảnh an toàn và khẩn cấp, tìm đường đi nhanh chóng và an toàn là quan trọng.

# 2.4. Mục tiêu và ý nghĩa
Mục tiêu của nhóm là giải quyết bài toán tìm đường trong mê cung và hiển thị kết quả một cách trực quan. Việc này không chỉ giúp hiểu rõ về cách các thuật toán tìm kiếm đồ thị hoạt động mà còn có thể ứng dụng trong các lĩnh vực thực tế như đã nêu ở trên. Kết quả của chương trình có thể cung cấp một giải pháp thực tế cho các vấn đề tương tự trong thế giới thực.


#**3. Phương pháp, công cụ sử dụng**
# 3.1. Các giải thuật tìm kiếm
Chương trình sử dụng ba thuật toán tìm kiếm đồ thị phổ biến để giải quyết bài toán tìm đường trong mê cung:

* BFS (Breadth-First Search): Thuật toán tìm kiếm theo chiều rộng, khám phá tất cả các ô láng giềng trước khi chuyển sang các ô láng giềng tiếp theo.

* DFS (Depth-First Search): Thuật toán tìm kiếm theo chiều sâu, tiếp tục đi sâu vào một nhánh của cây tìm kiếm trước khi quay lại và thử nhánh khác.

* A (A Star): Thuật toán tìm kiếm thông minh, sử dụng giá trị đánh giá (heuristic) để ưu tiên các ô có khả năng giúp tới mục tiêu.

# 3.2. Công cụ sử dụng
Chương trình sử dụng hai thư viện chính để đạt được mục tiêu của mình:

* Turtle Graphics: Dùng để vẽ và hiển thị đồ hoạ mê cung cũng như đường đi tìm được. Turtle là một thư viện đồ họa đơn giản và dễ sử dụng.

* PySimpleGUI: Sử dụng để tạo giao diện người dùng (GUI). Thư viện này cung cấp các công cụ để xây dựng giao diện người dùng một cách thuận tiện và nhanh chóng.

# 3.3. Lý do
* Turtle Graphics: Được chọn vì tính đơn giản và khả năng vẽ đồ họa trực quan. Turtle là một công cụ phù hợp cho việc hiển thị mê cung và đường đi trên một cửa sổ đồ họa.

* PySimpleGUI: Được chọn để tạo giao diện người dùng một cách dễ dàng. Việc này giúp người dùng có thể tương interact với chương trình, chẳng hạn như nhập mê cung từ tệp hoặc sử dụng giao diện để thiết lập mê cung.

# 3.4. Sử Dụng Cấu Trúc Dữ Liệu
* Danh sách và tập hợp: Dùng để lưu trữ thông tin về mê cung, đường đi, và các ô đã thăm.

* Deque (Double-ended Queue): Được sử dụng trong BFS để lưu trữ frontier, giúp quản lý các ô cần kiểm tra.

* Turtle Coordinates: Sử dụng để theo dõi và vẽ đường đi trên cửa sổ đồ họa.

# 3.6. Dataset

* Đầu ra (Output): Đường đi tìm được và hiển thị trực quan trên cửa sổ đồ họa.

```python
def create_maze_with_GUI():
    # Thiết lập giao diện để nhập mê cung bằng PySimpleGUI
    layout = [[sg.Text('Select how you want to create maze: ')],
              [sg.Button('Import File'), sg.Button('Use GUI')]
              ]
    window = sg.Window('Welcome', layout)
    while True:
        event, values = window.read()
        if event == sg.WINDOW_CLOSED:
            break
        if event == 'Use GUI':
            window.close()
            create_maze_with_GUI()
        elif event == 'Import File':
            import_layout = [[sg.Text('Choose file to import:')],
                             [sg.Input(), sg.FileBrowse()],
                             [sg.OK()]
                             ]
            window1 = sg.Window('Import File', import_layout)
            event1, values1 = window1.read()
            if event1 == 'OK':
                global input_grid
                filename = values1[0]
                with open(filename, 'r') as f:
                    input_grid = []
                    for line in f:
                        row = line.strip()
                        input_grid.append(row)
            window1.close()
            window.close()

def create_maze_with_GUI():
    # Thiết lập giao diện để nhập số hàng và số cột
    layout = [[sg.Text('Number of rows:'), sg.Input(size=(5, 1), key='-ROWS-')],
              [sg.Text('Number of columns:'), sg.Input(size=(5, 1), key='-COLS-')],
              [sg.Button('Submit')]]

    # Tạo cửa sổ để nhập số hàng và số cột
    window = sg.Window('Maze Setup', layout)

    while True:
        event, values = window.read()
        if event == sg.WINDOW_CLOSED or event == 'Submit':
            break

    window.close()

    # Lấy giá trị số hàng và số cột từ giao diện
    m = int(values['-ROWS-'])
    n = int(values['-COLS-'])

    # Thiết lập giao diện mê cung với các ô và các nút chức năng
    layout = [[sg.Button('', size=(2, 1), key=(i, j), pad=(0, 0), button_color=('white', 'white')) for j in range(n)] for i in range(m)]
    layout.append([sg.Button('Set Start'), sg.Button('Set End'), sg.Button('Set Wall')])
    layout.append([sg.Button('Submit')])

    # Tạo cửa sổ mới để thiết lập mê cung
    window = sg.Window('Maze Setup', layout)

    start_set = False
    end_set = False
    wall_set = True
    start = end = None

    start_existed = 0
    end_existed = 0

    while True:
        event, values = window.read()
        if event == sg.WINDOW_CLOSED:
            break

        if type(event) == tuple:
            button = window[event]
            # Thiết lập ô mê cung khi người dùng nhấp vào
            if wall_set:
                if button.ButtonColor[1] == 'white':
                    button.update(button_color=('white', 'black'))
                elif button.ButtonColor[1] == 'black':
                    button.update(button_color=('white', 'white'))
            elif start_set:
                if start:
                    start.update(button_color=('white', 'white'))
                button.update(button_color=('white', 'red'))
                start = button
            elif end_set:
                if end:
                    end.update(button_color=('white', 'white'))
                button.update(button_color=('white', 'purple'))
                end = button
        elif event == 'Set Start':
            start_set = True
            end_set = False
            wall_set = False
        elif event == 'Set End':
            end_set = True
            start_set = False
            wall_set = False
        elif event == 'Set Wall':
            wall_set = True
            start_set = False
            end_set = False
        elif event == 'Submit':
            global input_grid
            input_grid = []
            tmp = "+" * (n + 2)
            input_grid.append(tmp)
            # Tạo mê cung từ trạng thái của các ô trong giao diện
            for i in range(m):
                row = '+'
                for j in range(n):
                    button = window[(i, j)]
                    if button.ButtonColor[1] == 'black':
                        row += '+'
                    elif button.ButtonColor[1] == 'white':
                        row += ' '
                    elif button.ButtonColor[1] == 'red':
                        row += 's'
                        start_existed = 1
                    elif button.ButtonColor[1] == 'purple':
                        row += 'e'
                        end_existed = 1
                row += '+'
                input_grid.append(row)
            input_grid.append(tmp)

            # Kiểm tra xem người dùng đã đặt điểm xuất phát và kết thúc hay chưa
            if start_existed == 0:
                layout1 = [[sg.Text('Start point not set!')],
                            [sg.Button('OK')]
                            ]
                window1 = sg.Window('Warning', layout1)
                event1, values1 = window1.read()
                if event1 == 'OK':
                    window1.close()

            if end_existed == 0:
                layout1 = [[sg.Text('End point not set!')],
                            [sg.Button('OK')]
                            ]
                window1 = sg.Window('Warning', layout1)
                event1, values1 = window1.read()
                if event1 == 'OK':
                    window1.close()
            if start_existed == 1 and end_existed == 1:
                break

    window.close()
```

# **4.2. Mê cung với Turtle Graphics**


Mục tiêu của mã nguồn: Xây dựng và hiển thị mê cung sử dụng thư viện Turtle của Python.

Thư viện và giao diện:

*   Giao diện đồ họa: Mê cung được biểu diễn trực tiếp trên màn hình với các đối tượng Turtle có màu sắc khác nhau, giúp tạo ra một hình ảnh đồ họa sống động.

Các lớp và chức năng:
*   Lớp Maze: Đại diện cho các ô trong mê cung. Các ô này có màu trắng và được đánh dấu bởi dấu "+".
*  Lớp Green: Đại diện cho đích (điểm kết thúc) trong mê cung. Có màu xanh và được đánh dấu bởi ký tự "e".

*   Lớp Blue và Yellow: Đại diện cho đường đi có thể đi qua và đường đã đi qua. Có màu xanh lam và vàng tương ứng.

Hàm setup_maze:
* Mục Đích: Tạo và hiển thị mê cung trên màn hình dựa trên một lưới được cung cấp.
* Thực Hiện: Duyệt qua mỗi ô trong lưới, với mỗi ký tự trong ô, thiết lập màu sắc và vị trí cho các đối tượng Turtle tương ứng.

Hàm back_route(x, y):
* Hiển thị đường đi từ ô kết thúc đến ô bắt đầu bằng cách sử dụng thông tin được lưu trong "solution".

Biến quan trọng:
* walls: Danh sách chứa tọa độ các ô đại diện cho tường trong mê cung.
* path: Danh sách chứa tọa độ các ô đại diện cho đường đi có thể đi qua.

Mô tả chức năng:
* Hiển thị mê cung với tường, đường đi, điểm xuất phát (đỏ), và đích (xanh) trên màn hình.
* Lấy thông tin về các tường và đường đi từ lưới được cung cấp, giúp tạo ra mê cung có cấu trúc động.







```python
# Maze by Turtle
wn = turtle.Screen()
wn.bgcolor("black")
wn.title("Chương trình giải mê cung")
wn.setup(1300, 700)

# Biến toàn cục để lưu trữ tọa độ của điểm bắt đầu và điểm kết thúc
global start_x, start_y, end_x, end_y

# Lớp Turtle để vẽ tường mê cung
class Maze(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("white")
        self.penup()
        self.speed(0)

# Lớp Turtle cho sprite màu xanh (điểm kết thúc)
class Green(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("green")
        self.penup()
        self.speed(0)

# Lớp Turtle cho sprite màu xanh lam (được sử dụng trong quá trình hiển thị thuật toán tìm đường)
class Blue(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("blue")
        self.penup()
        self.speed(0)

# Lớp Turtle cho sprite màu đỏ (điểm bắt đầu)
class Red(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("red")
        self.penup()
        self.speed(0)

# Lớp Turtle cho sprite màu vàng (được sử dụng trong quá trình hiển thị quá trình lùi lại)
class Yellow(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("yellow")
        self.penup()
        self.speed(0)

# Lớp Turtle để vẽ đường mê cung
class Black(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("black")
        self.penup()
        self.speed(0)

# Chú thích: Đoạn mã trên định nghĩa các lớp Turtle khác nhau cho các phần tử khác nhau trong mê cung.

# Hàm để thiết lập mê cung dựa trên lưới được cung cấp
def setup_maze(grid):
    # Tính toán tọa độ để đặt mê cung ở giữa màn hình
    maze_width = len(grid[0]) * 24
    maze_height = len(grid) * 24

    screen_x_start = -maze_width / 2
    screen_y_start = maze_height / 2

    # Lặp qua từng ô trong lưới và thiết lập mê cung
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            character = grid[y][x]  # Lấy ký tự tại vị trí x và y trên lưới

            screen_x = x * 24 + screen_x_start
            screen_y = screen_y_start - y * 24
            wn_x = screen_x_start + screen_x
            wn_y = screen_y_start + screen_y
            maze.goto(wn_x, wn_y)

            # Nếu ký tự là "+", nó biểu thị một tường
            if character == "+":
                maze.goto(screen_x, screen_y)
                maze.stamp()
                walls.append((screen_x, screen_y))

            # Nếu ký tự là " " hoặc "e", nó biểu thị một đường đi trống hoặc điểm kết thúc
            if character == " " or character == "e":
                path.append((screen_x, screen_y))

            # Nếu ký tự là "e", nó biểu thị điểm kết thúc
            if character == "e":
                green.color("purple")
                green.goto(screen_x, screen_y)
                global end_x, end_y, start_x, start_y
                end_x, end_y = screen_x, screen_y
                green.stamp()
                green.color("green")

            # Nếu ký tự là "s", nó biểu thị điểm bắt đầu
            if character == "s":
                start_x, start_y = screen_x, screen_y
                red.goto(screen_x, screen_y)

# Hàm để kết thúc chương trình và đóng cửa sổ đồ họa Turtle
def end_program():
    wn.exitonclick()
    sys.exit()

# Hiển thị đường đi tối ưu đã tìm được
def back_route(x, y):
    yellow.goto(x, y)
    yellow.stamp()
    while (x, y) != (start_x, start_y):
        yellow.goto(solution[x, y])
        yellow.stamp()
        x, y = solution[x, y]
```

4.3.1. BFS

Khởi tạo:

* Tạo một hàng đợi (queue) frontier để theo dõi các ô tiếp theo cần kiểm tra.
* Tạo một từ điển solution để lưu trữ thông tin về các ô đã được thăm và đường đi từ ô đó đến ô trước đó.

Thêm điểm xuất phát vào hàng đợi:

* Bắt đầu từ điểm xuất phát (start_x, start_y), thêm nó vào hàng đợi frontier.
Lặp Qua Các Ô trong Hàng Đợi:

Lặp qua các ô trong hàng đợi frontier cho đến khi hàng đợi trống.
* Trong mỗi lần lặp, lấy ô tiếp theo từ đầu hàng đợi và gán giá trị (x, y) cho nó.

Kiểm tra các ô lân lận:

* Kiểm tra các ô lân cận của ô hiện tại (x, y) trong 4 hướng: trái, phải, lên và xuống.
* Nếu ô lân cận có thể đi qua (không phải là tường) và chưa được thăm:
 - Đánh dấu ô lân cận là đã thăm.
 - Thêm ô lân cận vào hàng đợi frontier.
 - Lưu thông tin về ô lân cận vào từ điển solution để theo dõi đường đi.

Hiển thị quá trình tìm kiếm:

* Trong mỗi bước, cập nhật màu của Turtle blue để đánh dấu ô đang được kiểm tra trong quá trình tìm kiếm.

Kiểm tra điểm đích:

* Trong mỗi bước, kiểm tra xem điểm đích (end_x, end_y) đã được thăm chưa.
* Nếu điểm đích đã được thăm, thoát khỏi vòng lặp.

Thông báo nếu không tìm thấy đường đi:

* Nếu điểm đích không được thăm và hàng đợi trống, hiển thị cảnh báo rằng không tìm thấy đường đi.

Hiển thị đường đi (nếu có):

* Nếu điểm đích đã được thăm, sử dụng từ điển solution để theo dõi ngược lại từ điểm đích đến điểm xuất phát và hiển thị đường đi bằng Turtle yellow.


```python
def bfs(x, y):
    frontier.append((x, y))  # Thêm điểm xuất phát vào hàng đợi frontier
    solution[x, y] = x, y  # Gán điểm xuất phát vào giải pháp

    while len(frontier) > 0:  # Lặp cho đến khi hàng đợi frontier trống
        time.sleep(0.1)
        x, y = frontier.popleft()  # Lấy điểm tiếp theo từ hàng đợi và gán vào x và y

        if (end_x, end_y) in visited:  # Kiểm tra điều kiện thoát khi điểm đích đã được thăm
            break

        # Kiểm tra các ô xung quanh và thêm vào hàng đợi nếu chưa được thăm
        if (x - 24, y) in path and (x - 24, y) not in visited:  # Kiểm tra ô bên trái
            cell = (x - 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x - 24, y))

        if (x, y - 24) in path and (x, y - 24) not in visited:  # Kiểm tra ô bên dưới
            cell = (x, y - 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y - 24))

        if (x + 24, y) in path and (x + 24, y) not in visited:  # Kiểm tra ô bên phải
            cell = (x + 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x + 24, y))

        if (x, y + 24) in path and (x, y + 24) not in visited:  # Kiểm tra ô bên trên
            cell = (x, y + 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y + 24))

        green.goto(x, y)  # Di chuyển đối tượng green đến điểm hiện tại và đánh dấu
        green.stamp()

    if (end_x, end_y) not in visited:  # Nếu điểm đích không được thăm, hiển thị thông báo
        unreachable = [[sg.Text('Không tìm thấy đường đi')]]
        window2 = sg.Window('Cảnh báo', unreachable)
        while True:
            event2, values2 = window2.read()
            if event2 == sg.WINDOW_CLOSED:
                break
        window2.close()
```

4.3.2. DFS

Khởi tạo:

* Tạo một hàng đợi (queue) frontier để theo dõi các ô cần kiểm tra theo thứ tự từng lớp.
* Tạo một tập hợp visited để lưu trữ các ô đã được kiểm tra.

Thêm điểm xuất phát vào hàng đợi :

* Bắt đầu từ điểm xuất phát (start_x, start_y), thêm nó vào hàng đợi frontier và đánh dấu là đã kiểm tra trong visited.

Lặp qua các ô trong hàng đợi:

* Lặp qua các ô trong hàng đợi frontier cho đến khi hàng đợi trống.
* Trong mỗi lần lặp, lấy ô tiếp theo từ đầu hàng đợi và gán giá trị (x, y) cho nó.

Kiểm tra các ô lân cận:

* Kiểm tra các ô lân cận của ô hiện tại (x, y) trong 4 hướng: trái, phải, lên và xuống.
* Nếu ô lân cận chưa được kiểm tra (không thuộc visited), thêm ô đó vào hàng đợi frontier và đánh dấu là đã kiểm tra trong visited.

Hiển thị quá trình tìm kiếm:

* Trong mỗi bước, cập nhật màu của Turtle blue để đánh dấu ô đang được kiểm tra trong quá trình tìm kiếm.

Kiểm tra điểm đích:

* Trong mỗi bước, kiểm tra xem điểm đích (end_x, end_y) đã được kiểm tra chưa.
* Nếu điểm đích đã được kiểm tra, thoát khỏi vòng lặp.

Thông báo nếu không tìm thấy đường đi :

* Nếu điểm đích không được kiểm tra và hàng đợi trống, hiển thị cảnh báo rằng không tìm thấy đường đi.

Hiển thị đường đi (Nếu có):

* Nếu điểm đích đã được kiểm tra, sử dụng thông tin từ hàng đợi và visited để xác định đường đi từ điểm đích đến điểm xuất phát và hiển thị đường đi bằng Turtle yellow.





```python
def dfs(x, y):
    frontier.append((x, y))  # Thêm điểm xuất phát vào ngăn xếp frontier
    solution[x, y] = x, y  # Gán điểm xuất phát vào giải pháp

    while len(frontier) > 0:  # Lặp cho đến khi ngăn xếp frontier trống
        time.sleep(0.1)
        x, y = frontier.pop()  # Lấy điểm trên cùng từ ngăn xếp và gán vào x và y

        if (end_x, end_y) in visited:  # Kiểm tra điều kiện thoát khi điểm đích đã được thăm
            break

        # Kiểm tra các ô xung quanh và thêm vào ngăn xếp nếu chưa được thăm
        if (x - 24, y) in path and (x - 24, y) not in visited:  # Kiểm tra ô bên trái
            cell = (x - 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x - 24, y))

        if (x, y - 24) in path and (x, y - 24) not in visited:  # Kiểm tra ô bên dưới
            cell = (x, y - 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y - 24))

        if (x + 24, y) in path and (x + 24, y) not in visited:  # Kiểm tra ô bên phải
            cell = (x + 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x + 24, y))

        if (x, y + 24) in path and (x, y + 24) not in visited:  # Kiểm tra ô bên trên
            cell = (x, y + 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y + 24))

        green.goto(x, y)  # Di chuyển đối tượng green đến điểm hiện tại và đánh dấu
        green.stamp()

    if (end_x, end_y) not in visited:  # Nếu điểm đích không được thăm, hiển thị thông báo
        unreachable = [[sg.Text('Không tìm thấy đường đi')]]
        window2 = sg.Window('Cảnh báo', unreachable)
        while True:
            event2, values2 = window2.read()
            if event2 == sg.WINDOW_CLOSED:
                break
        window2.close()
```

4.3.3. A*

Hàm đánh giá (Heuristic Function):


* h(x, y) có thể được hiểu là ước lượng chi phí từ ô có tọa độ (x, y) đến ô đích (end_x, end_y). Thông thường, hàm đánh giá được tính toán bằng cách sử dụng khoảng cách Euclidean  giữa hai điểm.

Giải thuật A (A-star):

Khởi tạo:

* Khởi tạo một hàng đợi ưu tiên (priority queue) open_set để theo dõi các nút chưa kiểm tra theo thứ tự ưu tiên (theo giá trị f(x) = g(x) + h(x)).
* Khởi tạo tập hợp closed_set để theo dõi các nút đã kiểm tra.

Khởi tạo nút xuất phát:

* Tạo một nút xuất phát và thêm nó vào open_set.

Lặp qua các nút trong open_set:

* Lặp qua các nút trong open_set cho đến khi nút đích được kiểm tra hoặc open_set trống.
* Lấy nút có giá trị f(x) thấp nhất từ open_set.
Kiểm tra nút đích:

* Nếu nút đích được kiểm tra, thoát khỏi vòng lặp và xây dựng đường đi.

Kiểm Tra Các Nút Lân Cận:

* Kiểm tra các nút lân cận của nút hiện tại.
* Tính toán chi phí thực tế từ nút xuất phát đến nút lân cận và ước lượng chi phí còn lại tới nút đích.
* Cập nhật giá trị g(x) và h(x) cho nút lân cận.
* Nếu nút lân cận chưa được kiểm tra hoặc có giá trị f(x) thấp hơn, cập nhật thông tin và thêm nút lân cận vào open_set.

Hiển Thị Đường Đi (Nếu Có):

* Nếu nút đích đã được kiểm tra, sử dụng thông tin từ closed_set để xây dựng đường đi từ điểm xuất phát đến điểm đích và hiển thị nó bằng Turtle yellow.

```python
def heuristic(x, y):
    return abs(x - end_x) * abs(x - end_x) + abs(y - end_y) * abs(y - end_y)


def aStar(x, y):
    open_set = set()
    closed_set = set()
    came_from = {}

    g_score = {(x, y): 0}
    h_score = {(x, y): heuristic(x, y)}
    f_score = {(x, y): h_score[(x, y)]}

    solution[x, y] = x, y
    open_set.add((x, y))

    while open_set:
        time.sleep(0.2)
        (a, b) = min(open_set, key=lambda node: f_score[node])

        if (a, b) == (end_x, end_y):
            break

        open_set.remove((a, b))
        closed_set.add((a, b))

        for (next_x, next_y) in [(a + 24, b), (a - 24, b), (a, b + 24), (a, b - 24)]:
            if (next_x, next_y) in path:
                if (next_x, next_y) in closed_set:
                    continue

                tentative_g_score = g_score[(a, b)] + 24

                if (next_x, next_y) not in open_set or tentative_g_score < g_score[(next_x, next_y)]:
                    came_from[(next_x, next_y)] = (a, b)
                    g_score[(next_x, next_y)] = tentative_g_score
                    h_score[(next_x, next_y)] = heuristic(next_x, next_y)
                    f_score[(next_x, next_y)] = g_score[(next_x, next_y)] + h_score[(next_x, next_y)]

                    solution[(next_x, next_y)] = a, b

                    if (next_x, next_y) not in open_set:
                        open_set.add((next_x, next_y))
                    blue.goto((next_x, next_y))
                    blue.stamp()

        green.goto(a, b)
        green.stamp()

    if (end_x, end_y) not in solution:
        unreachable = [[sg.Text('No path can be found')]]
        window2 = sg.Window('Warning', unreachable)
        while True:
            event2, values2 = window2.read()
            if event2 == sg.WINDOW_CLOSED:
                break
        window2.close()
```
