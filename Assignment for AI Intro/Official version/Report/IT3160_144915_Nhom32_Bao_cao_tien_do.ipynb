<h3>2. Phương pháp thực hiện đề tài</h3>
- Công cụ xây dựng giao diện trực quan cho bài toán:
    <p>Sau khi tìm hiểu, cân nhắc một số thư viện có khả năng phục vụ cho việc minh hoạ bài toán một cách trực quan, thân thiện với người sử dụng, nhóm quyết định chọn turtle để sử dụng cho vấn đề này. Turtle là module có khả năng thực hiện một số thao tác di chuyển, "vẽ" trên màn hình, phù hợp với người mới bắt đầu xây dựng giao diện trên Python. Để bắt đầu làm việc với turtle, ta cần import thư viện với tên tương ứng vào file mã nguồn. Bên cạnh đó, nhóm cũng sử dụng một vài thư viện khác có sẵn của hệ thống để phục vụ cho việc trực quan hoá bài toán trở nên có hiệu quả hơn.</p>

```python
import turtle
import time
import sys
from collections import deque

wn = turtle.Screen()
wn.bgcolor("black")
wn.title("Maze Solving Program")
wn.setup(1300, 700)

def endProgram():
    wn.exitonclick()
    sys.exit()
```

<p> Giao diện mê cung được xây dựng từ các bức tường - không thể đi xuyên qua, 2 điểm bắt đầu và kết thúc là điểm xuất phát và đích đến của mê cung. Tại thời điểm hiện tại, chương trình đang sử dụng mê cung đầu vào là các kí tự '+' - biểu thị cho tường, ' ' biểu thị cho đường đi, s và e lần lượt là hai điểm bắt đầu và xuất phát của mê cung.</p>
<p> Trên giao diện trực quan, tường được biểu thị bằng các ô vuông màu trắng, điểm xuất phát là ô vuông màu đỏ, điểm kết thúc là ô vuông màu tím. Trong quá trình thực hiện thuật toán, tác tử sẽ di chuyển và "tô màu" lên mê cung, thể hiện các nút đã xét được hiển thị bằng màu xanh lá cây, nút đang xét sẽ được hiển thị bằng màu xanh dương, và đường đi tìm thấy được biểu thị bằng màu vàng </p>

```python
class Maze(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("white")
        self.penup()
        self.speed(0)


# Path found by algorithm(s)
class Green(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("green")
        self.penup()
        self.speed(0)


class Blue(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("blue")
        self.penup()
        self.speed(0)


# Agent class
class Red(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("red")
        self.penup()
        self.speed(0)


class Yellow(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.shape("square")
        self.color("yellow")
        self.penup()
        self.speed(0)

grid = [
"+++++++++++++++",
"+s+       + +e+",
"+ +++++ +++ + +",
"+ + +       + +",
"+ +   +++ + + +",
"+ + + +   + + +",
"+   + +   + + +",
"+++++ +   + + +",
"+     +   +   +",
"+++++++++++++++",
]

maze = Maze()
red = Red()
blue = Blue()
green = Green()
yellow = Yellow()

walls = []
path = []
visited = set()
frontier = deque()
solution = {}

def setup_maze(grid):
    global start_x, start_y, end_x, end_y
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            character = grid[y][x]
            screen_x = -588 + (x * 24)
            screen_y = 288 - (y * 24)

            if character == "+":
                maze.goto(screen_x, screen_y)
                maze.stamp()
                walls.append((screen_x, screen_y))

            if character == " " or character == "e":
                path.append((screen_x, screen_y))

            if character == "e":
                green.color("purple")
                green.goto(screen_x, screen_y)
                end_x, end_y = screen_x, screen_y
                green.stamp()
                green.color("green")

            if character == "s":
                start_x, start_y = screen_x, screen_y
                red.goto(screen_x, screen_y)

def backRoute(x, y):
    yellow.goto(x, y)
    yellow.stamp()
    while (x, y) != (start_x, start_y):
        yellow.goto(solution[x, y])
        yellow.stamp()
        x, y = solution[x, y]
```

- Thuật toán tìm đường đi sử dụng trong chương trình:
    <p>Với đề tài tìm đường trong một mê cung cho trước, chương trình sử dụng một số thuật toán tìm kiếm để giải quyết vấn đề, bao gồm BFS, DFS và A*. Đây là các thuật toán cơ bản được sử dụng khi thực hiện nhiệm vụ tìm kiếm trong Trí tuệ nhân tạo. Trong đó, nhóm đã tìm hiểu và cài đặt 2 thuật toán BFS và DFS.</p>
    <p>Dưới đây là code minh hoạ cho thuật toán BFS:

```python
def bfs(x, y):
    frontier.append((x, y))
    solution[x, y] = x, y

    while len(frontier) > 0:
        time.sleep(0.1)
        x, y = frontier.popleft()

        if (x - 24, y) in path and (x - 24, y) not in visited:
            cell = (x - 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x - 24, y))

        if (x, y - 24) in path and (x, y - 24) not in visited:
            cell = (x, y - 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y - 24))
            #print(solution)

        if (x + 24, y) in path and (x + 24, y) not in visited:
            cell = (x + 24, y)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x + 24, y))

        if (x, y + 24) in path and (x, y + 24) not in visited:
            cell = (x, y + 24)
            solution[cell] = x, y
            blue.goto(cell)
            blue.stamp()
            frontier.append(cell)
            visited.add((x, y + 24))
        green.goto(x, y)
        green.stamp()
```

<p>Chương trình được thực hiện bằng cách gọi các hàm đã được định nghĩa ở trên và chọn thuật toán muốn thực hiện.</p>

```python
setup_maze(grid)
bfs(start_x, start_y)
backRoute(end_x, end_y)
wn.exitonclick()
```
